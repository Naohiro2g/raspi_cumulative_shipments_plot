```python
def pre_process():
    date_values = (
        "2012/02/29",  0,
        "2012/12/03", 840000,
        "2013/01/14", 1000000,
        "2013/10/08", 1750000,
        "2013/11/13", 2000000,
        "2014/03/03", 2500000,
        "2014/06/11", 3000000,
        "2014/08/20", 3500000,
        "2014/10/13", 3800000,
        "2015/02/18", 5000000,
        "2015/12/10", 7000000,
        "2016/02/29", 8000000,
        "2016/09/09", 10000000,
        "2016/11/25", 11000000,
        "2017/03/18", 12500000,
        "2017/07/19", 14500000,
        "2017/12/20", 17000000,
        "2018/12/21", 23000000,
        "2019/03/01", 25000000,
        "2019/12/17", 30000000,
        "2021/01/21", 37000000,
        "2022/02/28", 46000000,
        "2024/05/14", 60000000,)

    dates = []
    values = []
    for item in date_values:
        if type(item) is str:
            dates.append(item)
        elif type(item) is int:
            values.append(item)
    return dates, values


```

```python
import datetime as dt
import matplotlib.pyplot as plt


def draw_graph(dates, values, language, limit_date=dt.date(2026, 12, 31), limit_units=70000000):

    if language == 'en':
        units_div = 1000000
        title = 'Raspberry Pi Cumulative Shipments (million units)'
        font_name = "Meiryo"
        title_font_size = 8
    elif language == 'ja':
        units_div = 1000000
        title = 'Raspberry Pi 累計出荷台数 (百万台)'
        font_name = "Meiryo"
        title_font_size = 10

    x = [dt.datetime.strptime(d, '%Y/%m/%d').date() for d in dates]
    y = [int(v) for v in values]


    fig = plt.figure(dpi=200, figsize=(3.5, 2.5))
    ax = fig.add_subplot(1, 1, 1)
    ax.set_xlim([dt.date(2012, 1, 1), limit_date])
    ax.set_ylim(0, limit_units)
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x / units_div))))

    plt.title(title, fontsize=title_font_size, fontname=font_name)
    plt.rcParams["font.family"] = font_name
    plt.rcParams["font.size"] = 7
    plt.rcParams['axes.grid'] = True
    plt.rcParams['axes.linewidth'] = 0.6
    plt.rcParams['grid.linestyle'] = '--'
    plt.rcParams['grid.linewidth'] = 0.3
    plt.rcParams['xtick.direction'] = 'in'
    plt.rcParams['ytick.direction'] = 'in'
    plt.rcParams['xtick.major.width'] = 1.2
    plt.rcParams['ytick.major.width'] = 1.2
    plt.plot(x, y,  linewidth=1.5,
             marker="D", markersize=2.5, markeredgewidth=0.5,
             markeredgecolor="green", markerfacecolor="#a08080")
    plt.show()

```

```python
dates, values = pre_process()
draw_graph(dates, values, 'ja')
draw_graph(dates, values, 'en')
```
